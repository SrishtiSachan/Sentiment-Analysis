```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()

data1= pd.read_csv('Train.csv')
data1.head()
```

```python
data1.isnull().values.any()
data1= data1.dropna()
data1=data1.reindex(range(27447))
data1
```

```python
#converting everyting to lower case
df= data1.copy()
df= df.apply(lambda x: x.astype(str).str.lower())

df['text']   
```

```python
import string
#symbols = "!\"#$%&()*+-./:;<=>?@[\]^_`{|}~\n"

#this is for removing punctutations
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
for i in range(27447):
    df['text'][i]= remove_punctuations(df['text'][i])

df['text']
```

```python
#this code is to remove any digit that may appear in the text.
for i in range(27447):
    df['text'][i] = ''.join([j for j in df['text'][i] if not j.isdigit()])
        
df['text']
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

df['sentiment']=df['sentiment'].map({'positive':0,'neutral':1,'negative':2})

# create the transform
vectorizer = TfidfVectorizer()
df= df.dropna()
# tokenize and build vocab
vectorizer.fit_transform(df['text'])
y= df['sentiment']
y
```

```python
vector = vectorizer.transform(df['text'].values)
vector.shape
X= vector
```

```python
#the first model is using svm 
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report,accuracy_score
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=2)
model= SVC(kernel='linear')
model.fit(X_train,y_train)
y_pred=model.predict(X_test)

print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
print(accuracy_score(y_test,y_pred))
```

```python
#this is logistic regression model
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(solver='liblinear').fit(X_train, y_train)
y_pred = clf.predict(X_test)

print(accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))

```

```python
#the next is doing logistic regression with cross validation
```

```python
from sklearn.model_selection import cross_val_score, KFold
logreg = LogisticRegression(solver= 'lbfgs',max_iter=10000,random_state=2)
print(cross_val_score(logreg, X_train, y_train, cv=5, scoring='accuracy').mean())
```
